```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
#Import dataset
DelayEvents = pd.read_csv('DelayEvents.csv')
DelayEvents.dropna(inplace=True)
```

```python
DelayEvents.head()
```

## Dump useless variables:

```python
I_discard = ['DelayEventID', 'Line', 'PossibleTweetsAboutThisTripBSMSNo']
#Matt_discard = ['avg_Temp', 'avg_Humidity', 'avg_Visibility', 'avg_WindSpeed', 'EndsWithTimingPoint', 'DwellTimeOutliers', 'ArriveDelayOutliers', 'DepartureDelayOutliers',
       #'ArriveLoadOutliers', 'OnsLoadOutliers', 'OffsLoadOutliers',
       #'LeaveLoadOutliers', 'StopToStopOutliers',
       #'PossibleTweetsAboutThisTripsLine', 'PossibleTweetsAboutThisTripBSMSNo']
#columns_discard = I_discard + Matt_discard
columns_discard = I_discard
```

```python
DelayEvents.drop(columns_discard,axis =1, inplace=True)
```

## Turn 'OperationDate' into a numeric field:

```python
#Break OperationDate into [year, month, day]
date_sep = DelayEvents['OperationDate'].str.split("-", n=2, expand=True)

print (date_sep[0].nunique(), date_sep[1].nunique())
```

This shows that the data belongs to one year and four different months. So, we can safely get rid of the 'year' information. We are going to record the date as month + day/(30 or 31). Months Sep, Nov (9, 11) had 30 days and Oct, Ded (10, 12) had 31 days.

```python
date_sep = date_sep.astype(float)
```

```python
DelayEvents['OperationDate'] = date_sep[1]+ np.around(date_sep[2]/(31- date_sep[1] % 2), decimals=2)
```

```python
DelayEvents.head()
```

## Turn 'Trip' into workable fields

```python
trip_sep = DelayEvents['Trip'].str.split("-", n=2, expand=True)
trip_sep;
trip_sep[1].value_counts();
```

```python
trip_time = trip_sep[2].str.split(":", n=1, expand=True).astype(float)

DelayEvents['Time'] = np.around(trip_time[0] + trip_time[1]/60, decimals=1)

DelayEvents['Direction'] = trip_sep[1].str.get(1)
```

```python
DelayEvents.drop('Trip',axis =1, inplace=True)
```

In case you want to get the first two lettres as 'Direction', take this:
trip_sep[1].str.get(1) + trip_sep[1].str.get(2)

Although I think what really matters is simply the first letter, because by far the vast majority is either of 'EB', 'WB', 'NB', 'SB', as seen below:

```python
(trip_sep[1].str.get(1) + trip_sep[1].str.get(2)).value_counts()
```

# Get dummy variables for categorical features

We consider Direction_S = 1 as Direction_N = -1. Similarly, for Direction_W, Direction_E.

```python
DelayEvents = pd.get_dummies(DelayEvents, columns=['DayType', 'Direction'])
DelayEvents.loc[DelayEvents.Direction_S == 1, 'Direction_N'] = np.array(-1)
DelayEvents.loc[DelayEvents.Direction_W == 1, 'Direction_E'] = np.array(-1)
DelayEvents.drop(['Direction_S', 'Direction_W'], axis=1, inplace=True)
```

```python
DelayEvents.tail()
```

```python
DE_c = DelayEvents.corr()

DE_c
```

```python
n_features = DelayEvents.shape[1]
corrolated = []
for i in range(n_features):
    for j in range(n_features):
        if i < j:
            if np.abs(DE_c.iloc[i, j]) >= 0.95:
                corrolated.append((i, j))
corrolated
```

```python
columns_to_ignore = [DelayEvents.columns[i] for i in [23, 8, 14, 18, 15, 19, 28, 31, 36, 40]]

DelayEvents.drop (columns=columns_to_ignore, inplace=True)

DelayEvents.head()
```
