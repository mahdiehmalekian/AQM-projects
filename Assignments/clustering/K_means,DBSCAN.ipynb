## 1.1  K-means and K-means++

```python
def K_means (X, k):
    """Returns a clustering using K_means algorithm.
    The output is in a dictionary format where keys are clusters and
    key values are lists consisting of indices of elements of X belonging to a key"""
    
    n = len(X)
    if n < 1: return
    if k < 1 or not isinstance (k, int):
        print ('k must be a positive integer')
        return
    
    if k == 1: return {k: [i for i in range(n)]}
    
    #Feature Scaling
    sc_X=StandardScaler()
    Z = sc_X.fit_transform(X)
    
    #Form initial cluster
    cluster = {i:np.random.choice(k) for i in range(n)}
    new_cluster = {}

    while True:
        #Find centroid for each cluster
        centroid = {}
        #Put the empty clusters in out
        out = []
        for j in range(k):
            if not j in out:
                cluster_j = [i for i in cluster if cluster[i] == j]
                if cluster_j == []: out.append(j)
                else:
                    centroid[j] = np.mean(Z[cluster_j], axis=0)


        #Form new cluster
        new_cluster = {}
        for i in range(n):
            #find Euclidian distance of Z[i] from j-th centroid
            distance = [np.sum(np.square(Z[i] - centroid[j])) for j in range(k) if not j in out]
            new_cluster[i] = distance.index(min(distance))
        
        if cluster == new_cluster: break
        cluster = new_cluster
        
    return cluster
```

```python
def K_means_pp(X, k):
    """Returns a clustering using K_means++ algorithm.
    The output is in a dictionary format where keys are clusters and
    key values are lists consisting of indices of elements of X belonging to a key"""
    
    n = len(X)
    if n < 1: return
    if k < 1 or not isinstance (k, int):
        print ('k must be a positive integer')
        return
    
    #Feature Scaling
    sc_X=StandardScaler()
    Z = sc_X.fit_transform(X)
    
    #Choose initial centers and form initial cluster
    first_center = np.random.choice(n)
    centers = [first_center]
    cluster = {i:0 for i in range(n)}
    #D keeps track of the minimum Euclidean distance of a non-center data point to the closest center
    D = {i:np.sum(np.square(Z[i] - Z[first_center])) for i in range(n) if not i in centers}
    while len(centers) < k:
        prob = [D[i]**2 for i in D]
        prob /= np.sum(prob)
        new_center = np.random.choice([i for i in D], p=prob)
        centers.append(new_center)
        cluster[new_center] = len(centers)-1
        del D[new_center]
        for data_point in D:
            d_new = np.sum(np.square(Z[data_point] - Z[new_center]))
            if d_new < D[data_point]:
                D[data_point] = d_new
                cluster[data_point] = len(centers)-1
                
    new_cluster = {}

    while True:
        #Find centroid for each cluster
        centroid = {}
        for j in range (k):
            cluster_j = [i for i in cluster if cluster[i] == j]
            centroid[j] = np.mean(Z[cluster_j], axis=0)

        #Form new cluster
        new_cluster = {}
        for i in range(n):
            #find Euclidian distance of Z[i] from j-th centroid
            distance = [np.sum(np.square(Z[i] - centroid[j])) for j in range(k)]
            new_cluster[i] = distance.index(min(distance))
        
        if cluster == new_cluster: break
        cluster = new_cluster
        
    return cluster
```

## 1.2 DBSCAN

```python
def eps_nbr (c, X, eps):
    """Finds the indices of points in X in eps-neighborhood of X[c]"""
    
    return [i for i in range(X.shape[0]) if np.sum(np.square(X[c] - X[i])) < eps]
```

```python
def treat_core(c, nbr_c, label, cluster, core, not_visited, X, eps, min_samples):
    """Appends c to core and assigns label to all neighbors of c.
    Also, it checks if neighbors of c are cores themselves, and if so, the function calls itself for the new core."""
    
    core.append(c)
    
    for i in nbr_c:
        cluster [i] = label
        if i in not_visited:
            nbr_i = eps_nbr (i, X, eps)
            not_visited.remove(i)
            if len(nbr_i) >= min_samples:
                nbr_x = [y for y in nbr_i if not y in core]
                treat_core(i, nbr_i, label, cluster, core, not_visited, X, eps, min_samples)
```

```python
def DBSCAN (X, eps=0.5, min_samples=5):
    """Returns a clustering using DBSCAN algorithm.
    The output is in a dictionary format where keys are clusters and
    key values are lists consisting of indices of elements of X belonging to a key"""
    
    cluster, core = {}, []
    not_visited = [i for i in range(X.shape[0])]
    label = -1
    
    while len(not_visited) > 0:
        c = np.random.choice(not_visited)
        not_visited.remove(c)
        nbr_c = eps_nbr(c, X, eps)
        if len(nbr_c) >= min_samples:
            label += 1
            treat_core(c, nbr_c, label, cluster, core, not_visited, X, eps, min_samples)
            
    noise = {i: -1 for i in range(X.shape[0]) if not i in cluster}

    return {**cluster, **noise}
```

## 1.3 Drawing the clusters

```python
def draw_cluster (X, cluster):
    """Takes a cluster in the form of a dictionary where keys are indices of elements of X and
    key value is the cluster to which the key belongs.
    Provides a scatter plot of elements with different clusters having different colors.
    It regards an index assigned to cluster -1 as noise and displays it as a black star."""
    
    
    L = [cluster[c] for c in cluster]
    k = len(list(set(L)))
    
    clusters = [[i for i in cluster if cluster[i] == j] for j in range(k)] 
    noise = [i for i in cluster if cluster[i] == -1]

    colors = cm.rainbow(np.linspace(0, 1, k))
    
    for cluster_j, c in zip(clusters, colors):
        plt.scatter(X[cluster_j, 0], X[cluster_j, 1], color=c, cmap='viridis')
    plt.scatter(X[noise, 0], X[noise, 1], color='black', marker='*')
```

## 2.2 Find the right number of clusters

The choice of $k =5$ above was arbitrary. The visualization of data suggests two main clusters. But, let's take the time to choose the right $k$ using silhouette analysis.

```python
#Assign instances to clusters randomly
L = [np.random.choice(10) for i in range(X.shape[0])]
y = np.array(L)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
# Code taken from https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

Diagrams suggest $k =2$ or $3$ or $4$, because in these cases no or very little instances have negative silhouette coefficients. On the other hand, silhouette score for $k=2$ is significantly larger than those of $k =3$ or 4. Therefore, $k=2$ is the best choice. We now go back to experimenting the home-brew code with $k=2$.
